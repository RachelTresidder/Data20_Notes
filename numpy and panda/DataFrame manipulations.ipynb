```python
import seaborn as sns

#seaborn = package with sample datasets and data visualisation stuff
```

```python
flights = sns.load_dataset('flights')
flights
```

```python
# this data is in long format - can pivot to wide
```

```python
flights_piv = flights.pivot('year', 'month', 'passengers')
# in () = index, then columns, then values

flights_piv
```

```python
# we can rearrange pivoted data - stack and unstack
# stack = multi-leveled index

flights_stack = flights_piv.stack()
flights_stack
```

```python
# data now have a multi-layered index - year + month
```

```python
# unstack

flights_stack.unstack()
```
