```python
import seaborn as sns
import pandas as pd
import numpy as np

titanic = sns.load_dataset('titanic')
titanic
```

```python
# drop deck - not needed

titanic = titanic.drop(['deck'], axis = 1)
titanic
```

```python
# also dropping sibsp, parch, pclass(duplicate of class and class is easier to understand), who and adult_male(duplicate of sex)
#     embarked(duplicate of embark_town)

titanic = titanic.drop(['sibsp', 'parch', 'who', 'adult_male', 'embarked', 'alone'], axis = 1)
```

```python
# decided to remove survived, pretty sure it's a duplicate of alive

titanic_dropped = titanic.drop(['survived'], axis = 1)
titanic_dropped
```

```python
# looking at data again
titanic_clean. head()
```

```python
whole_fare = titanic_clean['fare'].sum()
whole_fare
```

```python
# creating new  column and filling with 'fare/wholefare'

titanic_clean['fare % of total'] = titanic_clean.apply(lambda row: (round((row.fare/whole_fare), 4)*100), axis = 1)

titanic_clean
```
